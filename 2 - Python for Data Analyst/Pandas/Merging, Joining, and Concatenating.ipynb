There are 3 main ways of combining DataFrames together: Merging, Joining and Concatenating.

```python
import pandas as pd
```

### **Concatenating.**

Concatenation basically glues together `DataFrames`. Keep in mind that dimensions should match along the axis you are concatenating on. You can use `pd.concat` and pass in a list of `DataFrames` to concatenate together:

Example DataFrames:

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index = [0, 1, 2, 3])

df1
```

```python
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index = [4, 5, 6, 7])

df2
```

```python
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        index = [8, 9, 10, 11])

df3
```

```python
help(pd.concat)
```

```python
pd.concat([df1, df2, df3])
```

```python
pd.concat([df1, df2, df3], axis = 1)
```

### **Merging.**

The `merge` function allows you to merge DataFrames together using a similar logic as merging SQL Tables together. For example:

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

left
```

```python
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})   

right
```

```python
help(pd.merge)
```

```python
pd.merge(left, right, on = 'key')
```

`on` is the column to merge on.

There are four values that `how` can take:

+ `inner`: Inner join, keeping only rows with matching values in both DataFrames. This is the default value.

+ `outer`: Outer join, retains all rows from both DataFrames, and fills in NaN for mismatched values.

+ `left`: Left join, retains all rows from the left DataFrame and matching rows from the right `DataFrame`. Mismatched values from the right `DataFrame` will be filled with `NaN`.

+ `right`: Right join, retains all rows from the right DataFrame and matching rows from the left `DataFrame`. Mismatched values from the left `DataFrame` will be filled with `NaN`.

Or to show a more complicated example:

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})

print('left:\n', left)
    
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})


print('right:\n', right)
```

```python
pd.merge(left, right, how = 'outer', on = ['key1', 'key2'])
```

```python
pd.merge(left, right, how = 'inner', on = ['key1', 'key2'])
```

```python
pd.merge(left, right, how = 'left', on = ['key1', 'key2'])
```

```python
pd.merge(left, right, how = 'right', on = ['key1', 'key2'])
```

### **Joining.**

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index = ['K0', 'K1', 'K2']) 

print('left:\n', left)

right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index = ['K0', 'K2', 'K3'])

print('right:\n', right)
```

```python
left.join(right)
```

```python
right.join(left)
```

```python
left.join(right, how = 'outer')
```

```python
right.join(left, how = 'outer')
```

```python
left.join(right, how = 'inner')
```

```python
right.join(left, how = 'inner')
```

```python
left.join(right, how = 'left')
```

```python
left.join(right, how = 'right')
```
