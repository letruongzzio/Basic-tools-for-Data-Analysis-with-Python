Notice how the legend in our previous figure overlapped some of actual diagram. Can you figure out how to display the legend outside of the plot as shown below?

Try searching Google for a good stackoverflow link on this topic. If you can't find it on your own - [use this one for a hint.](http://stackoverflow.com/questions/23556153/how-to-put-legend-outside-the-plot-with-pandas)

```python
f = plt.figure()

df3.iloc[0:30].plot.area()
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```

`bbox_to_anchor` is a parameter in the `legend()` method of the `matplotlib.pyplot` library. It is used to determine the position of the legend in the graph.

The syntax of bbox_to_anchor is `(x, y)`, where `x` and `y` are the coordinate values of the annotation image. The `x` and `y` values range from 0 to 1, and are calculated from the lower left corner of the graph. For example, `(0.0, 0.0)` represents the lower left corner of the graph, `(1.0, 1.0)` represents the upper right corner of the graph.
