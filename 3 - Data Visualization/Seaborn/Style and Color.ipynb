We've shown a few times how to control figure aesthetics in `seaborn`.

Check out the documentation page for more info on these topics: https://stanford.edu/~mwaskom/software/seaborn/tutorial/aesthetics.html

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
tips.head()
```

### **Styles.**

```python
sns.countplot(data=tips, x='sex')
```

```python
help(sns.set_style)
```

```python
sns.set_style('ticks') # `style` must be one of white, dark, whitegrid, darkgrid, ticks
sns.countplot(data=tips, x='sex')
```

```python
sns.set_style('whitegrid')
sns.countplot(data=tips, x='sex')
```

### **Spine Removal.**

```python
help(sns.despine)
```

```python
sns.countplot(data=tips, x='sex')
sns.despine()
```

```python
sns.countplot(data=tips, x='sex')
sns.despine(left=True, bottom=True)
```

### **Height and Aspect.**

You can use `matplotlib`'s `plt.figure(figsize=(width, height))`  to change the size of most seaborn plots.

You can control the size and aspect ratio of most `seaborn` grid plots by passing in parameters: `height`, and `aspect`. For example:

```python
# Non Grid Plot
plt.figure(figsize=(12,3))
sns.countplot(x='sex', data=tips)
```

```python
# Grid Type Plot
sns.lmplot(x='total_bill', y='tip', height=2, aspect=4, data=tips)
```

### **Scale and Context.**

The `set_context()` allows you to override default parameters:

```python
help(sns.set_context)
```

```python
sns.set_context(context='poster', font_scale=1)
sns.countplot(x='sex', data=tips, palette='coolwarm')
```
