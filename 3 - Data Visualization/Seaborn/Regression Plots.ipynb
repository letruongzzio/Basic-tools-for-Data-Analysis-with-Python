`seaborn` has many built-in capabilities for regression plots, however we won't really discuss regression until the Machine Learning section of the course, so we will only cover the `lmplot()` function for now.

`lmplot()` allows you to display linear models, but it also conveniently allows you to split up those plots based off of features, as well as coloring the hue based off of features.

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

### **`lmplot()`.**

```python
help(sns.lmplot)
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip')
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm')
```

### **Working with Markers.**

```python
sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex', markers=['o', 'v'])
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex', markers=['o', 'v'], scatter_kws={'s': 100}) 
# `scatter_kws` is a dictionary that allows us to pass in arguments to the scatter plot
# `s` is the size of the markers
```

### **Using a Grid.**

We can add more variable separation through columns and rows with the use of a grid. Just indicate this with the `col` or `row` arguments:

```python
sns.lmplot(data=tips, x='total_bill', y='tip', col='sex')
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip', row='sex', col='time')
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip', row='sex', col='time', hue='sex', palette='coolwarm')
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip', col='day', hue='sex', palette='coolwarm')
```

### **Aspect and Size.**

`seaborn` figures can have their height and aspect ratio adjusted with the `height` and `aspect` parameters:

```python
sns.lmplot(data=tips, x='total_bill', y='tip', col='day', hue='sex', palette='coolwarm', aspect=0.6, height=8)
# `aspect` is the aspect ratio
# `height` is the height of each facet
```
