### **Distribution Plots.**

Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:

+ distplot
+ jointplot
+ pairplot
+ rugplot
+ kdeplot

Import:

```python
import seaborn as sns
```

### **Data.**

`seaborn` comes with built-in data sets!

`seaborn` data: https://github.com/mwaskom/seaborn-data

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

### **`distplot`.**

The `distplot` shows the distribution of a set of univariate observations.

```python
help(sns.distplot)
```

```python
sns.distplot(tips['total_bill'])
```

The line is a kernel density estimation (KDE) plot. If you want to remove it, then you should do the following:

```python
sns.distplot(tips['total_bill'], kde = False)
```

```python
sns.distplot(tips['total_bill'],kde = False, bins = 100) # bins = number of bars
```

### **`jointplot`.**

`jointplot` allows you to basically match up two distplots for bivariate data. With your choice of what `kind` parameter to compare with: `scatter`, `reg`, `resid`, `kde`, or `hex`.

```python
help(sns.jointplot)
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'scatter')
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'hex')
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'reg')
```

### **`pairplot`.**

`pairplot` will plot pairwise relationships across an entire dataframe (for the numerical columns) and supports a color hue argument (for categorical columns). 

```python
help(sns.pairplot)
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue = 'sex', palette = 'coolwarm')
# `hue` is used to color the data points based on the column name provided
# `palette` is used to color the data points based on the color provided
```

### **`rugplot`.**

`rugplot` are actually a very simple concept, they just draw a dash mark for every point on a univariate distribution. They are the building block of a KDE plot:

```python
sns.rugplot(tips['total_bill'])
```

### **`kdeplot`.**

`kdeplot` are [Kernel Density Estimation plots](http://en.wikipedia.org/wiki/Kernel_density_estimation#Practical_estimation_of_the_bandwidth). These KDE plots replace every single observation with a Gaussian (Normal) distribution centered around that value. For example:

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['tip'])
sns.rugplot(tips['tip'])
```
